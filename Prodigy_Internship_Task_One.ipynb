```python
# Filter data by gender
male_age = data[data['Sex'] == 'male']['Age']
female_age = data[data['Sex'] == 'female']['Age']

# Plotting histograms side by side
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

# Histogram for Male
axes[0].hist(male_age, bins=30, color='skyblue', alpha=0.7, edgecolor='black')
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Frequency')
axes[0].set_title('Age Distribution - Male')

# Histogram for Female
axes[1].hist(female_age, bins=30, color='lightpink', alpha=0.7, edgecolor='black')
axes[1].set_xlabel('Age')
axes[1].set_ylabel('Frequency')
axes[1].set_title('Age Distribution - Female')

plt.tight_layout()  # Adjust layout for better presentation
plt.show()
```
