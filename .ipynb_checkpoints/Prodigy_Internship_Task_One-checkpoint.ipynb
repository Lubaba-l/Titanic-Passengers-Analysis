```python
import pandas as pd
import matplotlib.pyplot as plt
# Load the dataset
data = pd.read_csv('test.csv')

# Display the first few rows to understand the structure of the data
data.head()
```

```python
# Define age groups
bins = [0, 18, 30, 50, 100]  # Adjust bins as needed
labels = ['0-17', '18-29', '30-49', '50+']

# Create a new column 'AgeGroup' based on age bins in the 'data' DataFrame
data['AgeGroup'] = pd.cut(data['Age'], bins=bins, labels=labels)
```

```python
import seaborn as sns

# Set the style of the plot
sns.set(style="whitegrid")

# Create a count plot for AgeGroup
plt.figure(figsize=(8, 6))
sns.countplot(x='AgeGroup', data=data, palette='viridis')

# Adding labels and title
plt.xlabel('Age Groups')
plt.ylabel('Frequency')
plt.title('Distribution of Age Groups')

# Show plot
plt.xticks(rotation=45)  # Rotate x-axis labels if needed
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
```

```python
# Filter data by gender
male_age = data[data['Sex'] == 'male']['Age']
female_age = data[data['Sex'] == 'female']['Age']

# Plotting histograms side by side
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

# Histogram for Male
axes[0].hist(male_age, bins=30, color='skyblue', alpha=0.7, edgecolor='black')
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Frequency')
axes[0].set_title('Age Distribution - Male')

# Histogram for Female
axes[1].hist(female_age, bins=30, color='lightpink', alpha=0.7, edgecolor='black')
axes[1].set_xlabel('Age')
axes[1].set_ylabel('Frequency')
axes[1].set_title('Age Distribution - Female')

plt.tight_layout()  # Adjust layout for better presentation
plt.show()
```
